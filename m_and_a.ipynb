```python
import pandas as pd
from datetime import datetime


df = pd.read_csv(r'only_wood_customer_us_1.csv')

headers = ["Gender", "FirstName", "LastName", "Email", "Age", "City", "Country", "Created_at", "Referral"]

#delete a column from df
df.drop(columns=["UserName"], inplace=True)
# Add a new column to df
df["Created_at"] = " "
df["Referral"] = " "

# Update the column headers of df
df.columns = headers
df.head()
```

```python
def df1(filename, headers, cols_to_delete):
    # Load the data from the file
    df = pd.read_csv(filename)
    
    # Delete the specified columns from df
    df.drop(columns=cols_to_delete, inplace=True)
    
    # Add new columns to df
    now = datetime.now()
    dates = now.date()
    df["Created_at"] = dates
    df["Referral"] = "1001"
    df["Country"] = "U.S.A"
    
    # Update the column headers of df
    df.columns = headers
    
    return df
```

```python
filename = 'only_wood_customer_us_1.csv'
headers = ["Gender", "FirstName", "LastName", "Email", "Age", "City", "Country", "Created_at", "Referral"]
cols_to_delete = ["UserName"]

df1 = df1(filename, headers, cols_to_delete)
df1['FirstName'] = df1['FirstName'].str.replace('[^a-zA-Z0-9 ]', '', regex=True)
df1['LastName'] = df1['LastName'].str.replace('[^a-zA-Z0-9 ]', '', regex=True)
df1['Gender'].replace({'0': 'Male', '1': 'Female', 'M': 'Male', 'F': 'Female'}, inplace=True)
df1.head()
```

```python
import csv
from io import StringIO

with open("only_wood_customer_us_3.csv") as file:
    csv_reader = file.read()
    
    csv = StringIO(csv_reader)
    IO_obj = csv.getvalue()
    data3 = IO_obj.split('\n')
    data_list = []

    for line_num, i in enumerate(data3):
        if line_num > 0:
            if len(i)>0:
                values = i.split('\t')

                if 'string_' in values[0]:
                    gender=values[0].split('string_')[1]
                    
                if 'boolean_' in values[0]:
                    gender=values[0].split('boolean_')[1]
                    
                if 'character_' in values[0]:
                    gender=values[0].split('character_')[1]
                
                name = values[1].split('string_')[1]

                if values[2] =='N/A':
                    email=values[2]
                else:
                    email = values[2].split('string_')[1].lower()
                
                age = values[3].split('integer_')[1]

                city = values[4].split('string_')[1].title()

                country = values[5].split('string_')[1].title()

                data=[gender, name, email, age, city, country]
                
                data_list.append(data)

headers = ['Gender', 'Name', 'Email', 'Age', 'City', 'Country']

df3 = pd.DataFrame(data_list, columns=headers)
df3['Name'] = df3['Name'].str.replace('[^a-zA-Z0-9 ]', '', regex=True)
df3[['FirstName', 'LastName']] = df3['Name'].str.split(' ', expand=True)
df3.drop('Name', axis=1, inplace=True)
df3['Age'] = df3['Age'].str.replace('[^0-9]+', '', regex=True)
df3['Gender'].replace({'0': 'Male', '1': 'Female', 'M': 'Male', 'F': 'Female'}, inplace=True)
df3['Country'] = "U.S.A"
now = datetime.now()
dates = now.date()
df3["Created_at"] = dates
df3["Referral"] = "1003"
# Define a list of header names in the order you want them to appear
headers = ["Gender", "FirstName", "LastName", "Email", "Age", "City", "Country", "Created_at", "Referral"]

# Use the .loc accessor to rearrange the headers
df3 = df3.loc[:, headers]
df3.head(20)
              
```

```python
def my_m_and_a(df1, df2, df3):

    # Concatenate the dataframes vertically
    merged_df = pd.concat([df1, df2, df3], axis=0)

    # Return the merged dataframe as a CSV string
    return merged_df
```
